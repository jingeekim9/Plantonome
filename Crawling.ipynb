```python
import numpy as np
import pandas as pd
```

```python
import requests
#BeautifulSoup
from bs4 import BeautifulSoup

URL = "https://www.uniprot.org/uniprot/?query=taxonomy:33090&columns=entry name,reviewed,protein names,genes,organism,comment(FUNCTION),feature(HELIX),lineage(all),length&format=tab"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
data = r.content
data = data.decode('utf-8')
```

```python
cleaned_data = data.split("\n")
```

```python
for index, row in enumerate(cleaned_data):
    tempArray = row.split('\t')
    cleaned_data[index] = np.array(tempArray)
cleaned_data.pop()
```

```python
clean_numpy_array = np.array(cleaned_data)
```

```python
df = pd.DataFrame(clean_numpy_array)
```

```python
df = df.rename(columns=df.iloc[0])
df = df.drop(df.index[0])
```
